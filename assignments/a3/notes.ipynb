## Problem 1: 

**a)** The identity is resolved as follows: 

$$ 
\begin{align}
R_{\hat{n}}(\theta{}) &= e^{-i \frac{\theta{}}{2}\hat{n}\cdot{}\vec{\sigma{}}}                                             \tag{Given by (1)} \\
&\equiv{} \sum{}_{k = 0}^{\infty{}} \left(\frac{\left(i \frac{\theta}{2} \vec{\sigma_{n}}   \right)^{k}}{k!}    \right)      \tag{via the Maclaurin series expansion of $e^x$.} \\
&\equiv{} \sum{}_{k = 0}^{\infty{}} \left(\frac{\left(i \frac{\theta}{2} \vec{\sigma_{n}}   \right)^{(2k)}}{(2k)!}    \right) + \sum{}_{k = 0}^{\infty{}} \left(\frac{\left(i \frac{\theta}{2} \vec{\sigma_{n}}   \right)^{(2k+1)}}{(2k + 1)!}    \right)                                                  \tag{splitting into even and odd sums.} \\




\end{align}
$$

By Borns' Rule and Normalization rules, we know that $\vec{\sigma{}_n}^2 = 1$. (We also use the properties of $i$ to simplify further.) Therefore, we can simplify the Maclaurin series: 

$$
\begin{align}
&\equiv{} \sum{}_{k = 0}^{\infty{}} \left(\frac{\left((-1)^{k} \left(\frac{\theta}{2}\right)^{2k}    \right)}{(2k)!} \cdot I    \right) + \sum{}_{k = 0}^{\infty{}} \left(\frac{\left(-i (-1)^k \left(\frac{\theta}{2}\right)^{2k + 1}    \right)}{(2k + 1)!} \cdot{} \vec{\sigma_{n}}    \right)  \\
&\equiv{} \cos{\frac{\theta{}}{2}} \cdot{} I - i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n} \qquad{} \tag{By expressing the Maclaurin series in closed form.}
\end{align}

**b)** To prove that $R_{\hat{n}}(\theta{})$ is unitary, we must show that $R_{\hat{n}}(\theta{})^{\dagger} = R_{\hat{n}}(\theta{})^{-1}$, which implies that $R_{\hat{n}}(\theta{})^{\dagger} \cdot R_{\hat{n}}(\theta{}) = I$.

Hence, we must resolve the identity:

First, we have that $R_{\hat{n}}(\theta{})^{\dagger} = \left(\cos{\frac{\theta{}}{2}} \cdot{} I - i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n} \right)^{\dagger} = \cos{\frac{\theta{}}{2}} \cdot{} I + i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n}$.

This holds as both the identity and the Pauli matrices are Hermitian.

Next, we have that $R_{\hat{n}}(\theta{})^{\dagger} \cdot R_{\hat{n}}(\theta{}) = \left(\cos{\frac{\theta{}}{2}} \cdot{} I + i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n} \right) \cdot \left(\cos{\frac{\theta{}}{2}} \cdot{} I - i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n} \right) = \cos^{2}{\frac{\theta{}}{2}} \cdot{} I + \sin^{2}{\frac{\theta}{2}} \cdot{} I = I$.

As we have just resolved the identity, we have shown that $R_{\hat{n}}(\theta{})$ is unitary.

Next, we can show that $R_{\hat{n}}(\theta{})^\dagger = R_{\hat{n}}(-\theta{})$.

This may be achieved by subbing in $-\theta$ into the expression for $R_{\hat{n}}(\theta{})$. 

$$
\begin{align}
R_{\hat{n}}(-\theta{}) &= \cos{\frac{-\theta{}}{2}} \cdot{} I - i \sin{\frac{-\theta}{2}} \cdot \vec{\sigma_n} \\
&= \cos{\frac{\theta{}}{2}} \cdot{} I + i \sin{\frac{\theta}{2}} \cdot \vec{\sigma_n} \\
&= R_{\hat{n}}(\theta{})^{\dagger}
\end{align}
$$


## Problem 2:

To show that $R_x(\theta{}) \ket{0}$ produces a state corresponding to the Bloch vector $\vec{v}$ (as given on the problem set), we must resolve the action of $R_x(\theta{})$ on the state $\ket{0}$. 

First we recall the identity $R_x(\theta{}) = e^{-i \frac{\theta{}}{2} \sigma_x} = \cos{\frac{\theta{}}{2}} \cdot{} I - i \sin{\frac{\theta{}}{2}} \cdot{} \sigma_x$.

In matrix form, this is given by:

$$
\begin{align}
R_x(\theta{}) &= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & -i \sin{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} & \cos{\frac{\theta{}}{2}} \end{bmatrix} \\
\end{align}
$$

We can then apply $R_x(\theta{})$ to the state $\ket{0}$:

$$
\begin{align}
R_x(\theta{}) \ket{0} &= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & -i \sin{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} & \cos{\frac{\theta{}}{2}} \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} \\
&= \begin{bmatrix} \cos{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} \end{bmatrix} \\
\end{align}
$$

We then plot the derived state in the Bloch sphere.

$$
\begin{align}
\langle{} \vec{v} \rangle{} = \begin{bmatrix} (\bra{0} R_x(\theta{})) X (R_x(\theta{}) \ket{0}) \\ (\bra{0} R_x(\theta{})) Y (R_x(\theta{}) \ket{0}) \\ (\bra{0} R_x(\theta{})) Z (R_x(\theta{}) \ket{0}) \end{bmatrix}
\end{align}
$$

We resolve the above expression to find the Bloch vector $\vec{v}$. We can begin with the first component:

$$
\begin{align}
(\bra{0} R_x(\theta{})) X (R_x(\theta{}) \ket{0}) &= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & i\sin{\frac{\theta}{2}} \end{bmatrix} \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} \begin{bmatrix} \cos{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} \end{bmatrix} \\
&= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & -i \sin{\frac{\theta{}}{2}} \end{bmatrix} \begin{bmatrix}  -i \sin{\frac{\theta{}}{2}} \\ \cos{\frac{\theta{}}{2}} \end{bmatrix} \\
&= -i \sin{\frac{\theta{}}{2}} \cos{\frac{\theta{}}{2}} + i \sin{\frac{\theta{}}{2}} \cos{\frac{\theta{}}{2}} \\
&= 0
\end{align}
$$

We move to the second component: 

$$
\begin{align}
(\bra{0} R_x(\theta{})) Y (R_x(\theta{}) \ket{0}) &= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & i\sin{\frac{\theta}{2}} \end{bmatrix} \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix} \begin{bmatrix} \cos{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} \end{bmatrix} \\
&= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & i \sin{\frac{\theta{}}{2}} \end{bmatrix} \begin{bmatrix} - \sin{\frac{\theta{}}{2}} \\ i \cos{\frac{\theta{}}{2}} \end{bmatrix} \\
&= -\cos{\frac{\theta{}}{2}} \sin{\frac{\theta{}}{2}} + i \sin{\frac{\theta{}}{2}} i\cos{\frac{\theta{}}{2}} \\
&= -2 \cos{\frac{\theta{}}{2}} \sin{\frac{\theta{}}{2}} \\
&= -\sin{\theta{}}
\end{align}
$$

Finally, we resolve the third component:

$$
\begin{align}
(\bra{0} R_x(\theta{})) Z (R_x(\theta{}) \ket{0}) &= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & i\sin{\frac{\theta}{2}} \end{bmatrix} \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} \begin{bmatrix} \cos{\frac{\theta{}}{2}} \\ -i \sin{\frac{\theta{}}{2}} \end{bmatrix} \\
&= \begin{bmatrix} \cos{\frac{\theta{}}{2}} & i \sin{\frac{\theta{}}{2}} \end{bmatrix} \begin{bmatrix} \cos{\frac{\theta{}}{2}} \\ i \sin{\frac{\theta{}}{2}} \end{bmatrix} \\
&= \cos^{2}{\frac{\theta{}}{2}} + (i \sin{\frac{\theta{}}{2}})^{2} \\
&= cos^{2}{\frac{\theta{}}{2}} - \sin^{2}{\frac{\theta{}}{2}} \\
&= \cos{\theta{}}
\end{align}
$$

Therefore, the Bloch vector $\vec{v}$ is given by $\begin{bmatrix} 0 \\ -\sin{\theta{}} \\ \cos{\theta{}} \end{bmatrix}$, which is the same as the given Bloch vector in eqn. (4) in the problem set.
